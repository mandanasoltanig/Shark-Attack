```python
import pandas as pd

url= "D:/04_IronHack_DA/IRONHACK_LABS_PROJECTS/Shark-Attack/df_filtered.csv"
df = pd.read_csv(url, low_memory=False)
#df
```

```python
df.dropna(how="all", inplace=True) # dropped all the rows that were completely filled with nan vales

df.to_csv(r"D:/04_IronHack_DA/IRONHACK_LABS_PROJECTS/Shark-Attack/df_filtered.csv", index=False)
#df
```

```python
#column change --> Year

df["Date"]=df["Date"].fillna(0) #filled empty spaces with 0 as int


df["Year"]=df["Year"].fillna(0.0) #filled empty spaces with 0.0 as float
df["Year"]=df["Year"].astype(int) #converted the rows into type "int" to get rid of .0 at the end
df["Year"] = df["Year"].replace([1500,1000,77,5],0) #replaced the odd entries with 0 
df.to_csv(r"D:/04_IronHack_DA/IRONHACK_LABS_PROJECTS/Shark-Attack/df_filtered.csv", index=False)
df
```

```python
#column change --> Type


df["Type"]=df["Type"].replace({'unprovoked':'Unprovoked',
' Provoked':'Provoked',
'Unverified':'Unconfirmed',
'Questionable':'Unconfirmed',
'?':'Unconfirmed',
'Boat':'Watercraft'}) #replaced the duplicated and unnecessary enteries

df["Type"]=df["Type"].fillna("Invalid") #NAN values replaced with Invalid
df.to_csv(r"D:/04_IronHack_DA/IRONHACK_LABS_PROJECTS/Shark-Attack/df_filtered.csv", index=False)


#df["Type"].unique()
```

```python
#column change --> Country


df["Country"]=df["Country"].fillna("Unknown") #fills the blanks with/nan with "Unknown"
df["Country"] = df["Country"].str.strip().str.lower() #removes the extra space from begin and end. turns the country names to lower case
#df["Country"].value_counts
#print(df["Country"].value_counts())

df["Country"] = df["Country"].str.replace("/", " and ", regex=False) #removes the "/" between the names: andaman / nicobar islands --> andaman and nicobar islands

df["Country"]=df["Country"].replace({
    "south atlantic ocean":"Unknown", 
    "caribbean sea":"Unknown",
    "indian ocean":"Unknown",
    "north pacific ocean":"Unknown",
    "mid atlantic ocean":"Unknown",
    "north atlantic ocean":"Unknown",
    "central pacific":"Unknown",
    "southwest pacific ocean":"Unknown",
    "south pacific ocean":"Unknown",
    "mediterranean sea":"Unknown",
    "palestinian territories":"palestin",
    "red sea / indian ocean":"Unknown",
    "north sea":"Unknown",
    "red sea":"Unknown",
    "south china sea":"Unknown",
    "northern arabian sea":"Unknown",
    "bay of bengal":"Unknown",  
    "italy / croatia":"Unknown",
    "iran / iraq":"Unknown",
    "sudan?":"sudan",
    "mid-pacifc ocean":"Unknown",
    "ocean":"Unknown",
    "indian ocean?":"Unknown",
    "africa":"Unknown",
    "coast of africa":"Unknown",
    "tasman sea":"Unknown",
    "between portugal & india":"Unknown",
    "red sea?":"Unknown",
    "asia?":"Unknown",
    "ceylon (sri lanka)":"sri lanka",
    "british overseas territory":"Unknown",
    "united arab emirates (uae)":"united arab emirates",
    "st helena, british overseas territory":"st helena",
    "egypt  and  israel":"Unknown",
    "gulf of aden":"Unknown"
})

df.to_csv(r"D:/04_IronHack_DA/IRONHACK_LABS_PROJECTS/Shark-Attack/df_filtered.csv", index=False)
#print(df["Country"].value_counts().to_string())
#df["Country"].unique()

#for i in df["Country"]:
   # if i == "united arab emirates (uae)":
       # print(i)
```

```python
#column change --> State


df["State"]=df["State"].fillna("Unknown")
df["State"] = df["State"].str.strip().str.lower()
#df["State"].unique()
print(df["State"].value_counts().to_string())
df.to_csv(r"D:/04_IronHack_DA/IRONHACK_LABS_PROJECTS/Shark-Attack/df_filtered.csv", index=False)


```

```python
#column change --> Location
df["Location"]=df["Location"].fillna("Unknown")
df["Location"] = df["Location"].str.strip().str.lower()
print(df["Location"].value_counts().to_string())
df.to_csv(r"D:/04_IronHack_DA/IRONHACK_LABS_PROJECTS/Shark-Attack/df_filtered.csv", index=False)
```

```python
#column change --> Source

df["Source"]=df["Source"].fillna("Unknown")
df["Source"] = df["Source"].str.strip().str.lower()
df["Source"].value_counts().to_string()
```
