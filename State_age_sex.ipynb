```python
#Cleaning state column

#display(df['State'])#Full column

#display(df['State'].unique())#Unique States

#display(df['State'].value_counts(dropna=False))#Counting each value

#Removed Extra spaces
#df['State'] = df['State'].str.strip()
#df['State']

#Standardized Text
#df['State'] = df['State'].str.title()
#df['State']

#Removed Invalid non state entries

invalid_patterns = ["off","island","sea","ocean","gulf","bay","coast","unknown"]#List of unwanted words

pattern = "|".join(invalid_patterns)#combined them into one search pattern, "|" OR

df.loc[df['State'].str.contains(pattern, case=False, na=False), 'State'] = "Unknown" #Found those rows and removed it,checks Uppercase/Lowercase,skips missing value safely 

print(df['State'].value_counts(dropna=False).head(20))#In state count Each State appears,include NAN value
#display(df['State'].unique())

df.to_csv("/Users/akashkumarsamantray/Shark-Attack/df_filtered.csv",index=False)
```

```python
# Cleaning Age
#display(df['Age'])

#Replaced Messy values

df['Age_clean'] = df['Age_clean'].replace({
    # text labels
    'teen': 15, 'Teen': 15, 'Teens': 15,
    'Middle age': 45, '"middle-age"': 45,
    'mid-30s': 35, 'mid-20s': 25,
    'adult': 25, '(adult)': 25,
    '"young"': 20, 'young': 20,
    'Elderly': 70,
    'a minor': 10, '?': "Unknown", '!!': "Unknown", '!2': "Unknown", '!6': "Unknown",
    'M': "Unknown", 'F': "Unknown", 'X': "Unknown", 'A.M.': "Unknown",
    'MAKE LINE GREEN':"Unknown", '\xa0': "Unknown", '': "Unknown",
    '20/30': 25,
})
#print(df['Age_clean'].describe())
df['Age_clean'] = df['Age_clean'].astype('Int64')

print("Nulls:", df['Age_clean'].isna().sum())
df['Age_clean'].unique()


```

```python
# Sex Cleaning

#Strip White spaces
df['Sex_clean'] = df['Sex'].str.strip()
#df['Sex_clean']

# Uppercase m -> m
df['Sex_clean'] = df['Sex_clean'].str.upper()
#df['Sex_clean']

# Replace odd values
df['Sex_clean'] = df['Sex_clean'].replace({
    'M X 2': 'M',
    '?': 'Unknown',
    '.': 'Unknown',
    'N': 'Unknown',
    'LLI': 'Unknown',
    'UNKOWN': 'Unknown', 
})
df['Sex_clean'] = df['Sex_clean'].fillna('Unknown')#
#df['Sex_clean']

print(df['Sex_clean'].value_counts(dropna=False))
print("\nUnique:", df['Sex_clean'].unique())
```
