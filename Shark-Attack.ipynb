```python
import pandas as pd

url= "D:/04_IronHack_DA/IRONHACK_LABS_PROJECTS/Shark_Attack/GSAF5.csv"
df = pd.read_csv(url, low_memory=False)
#df
```

```python
df.columns=df.columns.str.strip()
df_filtered=df[["Date", "Year","Type","Country","State","Location","Activity","Name","Sex","Age","Injury","Fatal Y/N", "Time","Species","Source","pdf","href","Case Number","original order"]].copy()
df_filtered
```

```python
df_filtered.nunique()
#df_filtered.dtypes
```

```python
#cleaning the column --> Date

month_map = {
    "January": "Jan",
    "February": "Feb",
    "March": "Mar",
    "April": "Apr",
    "May": "May",
    "June": "Jun",
    "July": "Jul",
    "August": "Aug",
    "September": "Sep",
    "October": "Oct",
    "November": "Nov",
    "December": "Dec"
}
df_filtered["Date"]=df_filtered["Date"].str.replace("-"," ") #removes "-"
df_filtered["Date"]=df_filtered["Date"].str.replace("Reported"," ")#removes "reported"
df_filtered["Date"]=df_filtered["Date"].str.replace("`"," ")#removes "`"
df_filtered["Date"]=df_filtered["Date"].str.replace("."," ")#removes"."
df_filtered["Date"]=df_filtered["Date"].astype(str).replace(month_map, regex=True) #replaces june or march with jun or mar
df_filtered["Date"] = df_filtered["Date"].str.replace(r"(\d{1,2})(st|nd|rd|th)", r"\1", regex=True) #removes st, nd, rd, th 
df_filtered["Date"] = df_filtered["Date"].str.replace(r"^(\d)\s", r"0\1 ", regex=True) #adds 0 in the beginning --> 2 mar = 02 mar
df_filtered["Date"] = df_filtered["Date"].str.replace(r"\s+", " ", regex=True) #turns multiple spaces to single --> 02  mar = 02 mar
df_filtered["Date"] = df_filtered["Date"].str.replace( r"([A-Za-z]{3,})(\d{4})$", r"\1 \2", regex=True) #adds space between day,month, year --> 13 may2014 = 13 may 2014
df_filtered["Date"] = df_filtered["Date"].str.extract(r"^(\d{1,2}\s[A-Za-z]{3,})", expand=False) # extracts month and date from the enteries and makes the rest nan --> 02 may 2024 before summer = 02 may
df_filtered["Date"]=df_filtered["Date"].str.strip() # removes spaces from begin and end
df_filtered

df_filtered.to_csv(r"D:\04_IronHack_DA\IRONHACK_LABS_PROJECTS\Shark_Attack\df_filtered.csv", index=False)

#df_filtered.head(50)
#for i in df_filtered["Date"]:
 #   print(i)
```
